Fill the missing values using mean, median, ffill() and bfill() methods

```python
#find the missing values
df.isna().sum()
```

```python
#filling the Natural_Gas_Vol with its mean
df['Natural_Gas_Vol.'].fillna(df['Natural_Gas_Vol.'].mean(),inplace=True)
```

```python
#filling the Crude_oil_Vol. with its median
df['Crude_oil_Vol.'].fillna(df['Crude_oil_Vol.'].median(),inplace=True)
```

```python
#filling the Copper_Vol. with bfill method
df['Copper_Vol.'].fillna(method='bfill',inplace=True)
```

```python
#filling the platinum_Vol. with ffill method
df['Platinum_Vol.'].fillna(method='ffill',inplace=True)
```

```python
#filling the rest coloumns with bfill method
df=df.fillna(method='bfill')
```

```python
from statsmodels.tsa.arima.model import ARIMA
model=ARIMA(df['Apple_Price'],order=(1,1,1))
model_fit=model.fit()
print(model_fit.summary)
model_fit.plot_diagnostics(figsize=(7,6))
plt.show()
forecast=model_fit.forecast(steps=10)
forecast
```
